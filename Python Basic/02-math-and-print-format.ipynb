# String Formatting #

‡πÉ‡∏ä‡πâ method `format()` ‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡∏à‡∏±‡∏î‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏Ç‡∏≠‡∏á String

`"String with Placeholder".format(data1, data2)`

‡∏™‡πà‡∏ß‡∏ô‡πÉ‡∏´‡∏ç‡πà‡∏à‡∏∞‡πÉ‡∏ä‡πâ String Formatting ‡πÉ‡∏ô‡∏™‡πà‡∏ß‡∏ô **Output**

```python
name = "Alice"
age = 18
my_info = "My name is {}. I am {} years old. My height is {}".format(name, age, 1.82)
print(my_info)
```

### Empty Placeholder `{}` ###

`{}` ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ó‡∏ô‡∏î‡πâ‡∏ß‡∏¢‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏î ‡πÜ

* ‡∏î‡∏π‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡∏Ç‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏ô method `.format()`

```python
name = "Dad"
hour = 10
minute = 30
result = "Please contact {} at {}:{}".format(name, hour, minute)
print(result)
```

### Index Placeholder `{0}` ###

`{0}` ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ó‡∏ô‡∏î‡πâ‡∏ß‡∏¢‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏î ‡πÜ ‡πÇ‡∏î‡∏¢‡∏≠‡πâ‡∏≤‡∏á‡∏≠‡∏¥‡∏á‡∏•‡∏≥‡∏î‡∏±‡∏ö‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡∏Ç‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•

* ‡∏î‡∏π‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡∏Ç‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏ô method `.format()`

```python
name = "John"
age = 18
introducing = "My name is {0}, I am {1} years old.".format(name, age)
print(introducing)
introducing = "Yes, I am {1}. My name is {0}".format(name, age)
print(introducing)
```

### Placeholder `{:d}`, `{:f}`, `{:s}` ###

`{:d}` ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ó‡∏ô‡∏î‡πâ‡∏ß‡∏¢‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏ï‡πá‡∏° (**d**ecimal) <br>
`{:f}` ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ó‡∏ô‡∏î‡πâ‡∏ß‡∏¢‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏à‡∏£‡∏¥‡∏á (**f**loating point) <br>
`{:s}` ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ó‡∏ô‡∏î‡πâ‡∏ß‡∏¢‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏° (**s**tring) <br>

```python
sentence = "I got {:d} points in 01418{:d}.".format(80, 112)
print(sentence)
```

```python
sentence = "I got {:f} points in 01418{:d}.".format(80, 112)
print(sentence)
```

```python
sentence = "I got {:s} in {:s}.".format("A", "Fundamental Programming Concept")
print(sentence)
```

```python
grade = input("Enter your grade: ")
sentence = "I wish grade {:s} in 01418112".format(grade)
print(sentence)
```

### ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏Ñ‡∏ß‡∏≤‡∏°‡∏Å‡∏ß‡πâ‡∏≤‡∏á‡∏ó‡∏µ‡πà‡πÉ‡∏ä‡πâ‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡πÅ‡∏ó‡∏ô‡∏ó‡∏µ‡πà‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• ###

`{:[width][precision]}` `width` ‡πÄ‡∏õ‡πá‡∏ô‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏ï‡πá‡∏°

‡πÄ‡∏ä‡πà‡∏ô `{:10d}`, `{:20s}`, `{:12f}`

```python
grade = 'A'
sentence = "I wish grade {:3s} in 01418112".format(grade)
print(sentence)
print('-' * 27)
grade = 'B+'
sentence = "I wish grade {:2s} in 01418112".format(grade)
print(sentence)
```

```python
# ‡∏Å‡∏£‡∏ì‡∏µ‡πÑ‡∏°‡πà‡∏£‡∏∞‡∏ö‡∏∏ width
sentence1 = "{:s} {:f}".format("Lay's", 29)
sentence2 = "{:s} {:f}".format("Coca Cola", 19)
sentence3 = "{:s} {:f}".format("Super Size Icecream Bucket", -499)
print(sentence1)
print(sentence2)
print(sentence3)
```

```python
# ‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö‡∏Å‡∏±‡∏ö‡∏Å‡∏£‡∏ì‡∏µ‡∏£‡∏∞‡∏ö‡∏∏ width
sentence1 = "{:30s} {:12f}".format("Lay's", 29)
sentence2 = "{:30s} {:12f}".format("Coca Cola", 19)
sentence3 = "{:30s} {:12f}".format("Super Size Icecream Bucket", -499)
print(sentence1)
print(sentence2)
print(sentence3)
```

### Flag `<`, `>`, `^` ‡πÉ‡∏ô format ###

> Flag `<` ‡πÉ‡∏ä‡πâ‡∏à‡∏±‡∏î‡∏ä‡∏¥‡∏î‡∏ã‡πâ‡∏≤‡∏¢

```python
template = "{:<30s} {:<12f}"
sentence1 = template.format("Lay's", 29)
sentence2 = template.format("Coca Cola", 19)
sentence3 = template.format("Super Size Icecream Bucket", -499)
print(sentence1)
print(sentence2)
print(sentence3)
```

> ‡∏´‡∏≤‡∏Å‡πÑ‡∏°‡πà‡πÉ‡∏™‡πà flag, `{:[width]s}` ‡∏à‡∏∞‡∏à‡∏±‡∏î‡∏ä‡∏¥‡∏î‡∏ã‡πâ‡∏≤‡∏¢
>
> `{:[width]d}` ‡πÅ‡∏•‡∏∞ `{:[width]f}` ‡∏à‡∏∞‡∏à‡∏±‡∏î‡∏ä‡∏¥‡∏î‡∏Ç‡∏ß‡∏≤

```python
template = "{:30s} {:12f}"
sentence1 = template.format("Lay's", 29)
sentence2 = template.format("Coca Cola", 19)
sentence3 = template.format("Super Size Icecream Bucket", -499)
print(sentence1)
print(sentence2)
print(sentence3)
```

> Flag `>` ‡πÉ‡∏ä‡πâ‡∏à‡∏±‡∏î‡∏ä‡∏¥‡∏î‡∏Ç‡∏ß‡∏≤

```python
sentence1 = "{:>30s} {:>12f}".format("Lay's", 29)
sentence2 = "{:>30s} {:>12d}".format("Coca Cola", 19)
sentence3 = "{:>30s} {:>12f}".format("Super Size Icecream Bucket", -499)
print(sentence1)
print(sentence2)
print(sentence3)
```

> Flag `^` ‡πÉ‡∏ä‡πâ‡∏à‡∏±‡∏î‡∏Å‡∏∂‡πà‡∏á‡∏Å‡∏•‡∏≤‡∏á

```python
sentence1 = "[{:^33s}] [{:^20f}]".format("Lay's", 29)
sentence2 = "[{:^39s}] [{:^20d}]".format("Coca Cola", 19)
sentence3 = "[{:^30s}] [{:^20f}]".format("Super Size Icecream Bucket", -499)
print(sentence1)
print(sentence2)
print(sentence3)
```

### ‡πÉ‡∏ä‡πâ `.` ‡πÉ‡∏ô format `:f` ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏•‡∏Ç‡∏´‡∏•‡∏±‡∏á‡∏à‡∏∏‡∏î‡∏ó‡∏®‡∏ô‡∏¥‡∏¢‡∏° ###

* ‡πÉ‡∏ä‡πâ‡∏´‡∏•‡∏±‡∏Å‡∏Å‡∏≤‡∏£‡∏õ‡∏±‡∏î‡πÄ‡∏®‡∏©

```python
sentence1 = "{:>30s} {:12.2f}".format("Lay's", 29)
sentence2 = "{:>30s} {:12.3f}".format("Coca Cola", 19.5444)
sentence3 = "{:>30s} {:12.4f}".format("Super Size Icecream Bucket", -499.555555)
print(sentence1)
print(sentence2)
print(sentence3)
```

> ‡∏Å‡∏£‡∏ì‡∏µ‡πÄ‡∏®‡∏©‡∏ó‡∏µ‡πà‡∏õ‡∏±‡∏î‡πÄ‡∏õ‡πá‡∏ô‡πÄ‡∏•‡∏Ç 5 ‡∏à‡∏∞‡∏õ‡∏±‡∏î‡πÄ‡∏®‡∏©‡πÄ‡∏Ç‡πâ‡∏≤‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏π‡πà 
* ‡∏Å‡∏é‡∏ô‡∏µ‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏à‡∏£‡∏¥‡∏á‡πÄ‡∏°‡∏∑‡πà‡∏≠‡∏õ‡∏±‡∏î‡πÄ‡∏•‡∏Ç‡πÄ‡∏õ‡πá‡∏ô‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏ï‡πá‡∏°‡πÄ‡∏ó‡πà‡∏≤‡∏ô‡∏±‡πâ‡∏ô
* ‡πÉ‡∏ä‡πâ‡∏Å‡∏±‡∏ö‡∏Å‡∏≤‡∏£‡∏õ‡∏±‡∏î‡πÄ‡∏•‡∏Ç‡πÄ‡∏õ‡πá‡∏ô‡∏ó‡∏®‡∏ô‡∏¥‡∏¢‡∏°‡πÑ‡∏°‡πà‡πÑ‡∏î‡πâ‡πÄ‡∏™‡∏°‡∏≠‡πÑ‡∏õ ‡πÄ‡∏ô‡∏∑‡πà‡∏≠‡∏á‡∏à‡∏≤‡∏Å‡∏Å‡∏≤‡∏£‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏•‡∏Ç‡∏ó‡∏®‡∏ô‡∏¥‡∏¢‡∏°‡πÄ‡∏õ‡πá‡∏ô‡πÄ‡∏•‡∏Ç‡∏ê‡∏≤‡∏ô‡∏™‡∏≠‡∏á‡πÉ‡∏ô‡∏£‡∏∞‡∏ö‡∏ö‡∏Ñ‡∏≠‡∏°‡∏û‡∏¥‡∏ß‡πÄ‡∏ï‡∏≠‡∏£‡πå‡πÑ‡∏°‡πà‡πÅ‡∏°‡πà‡∏ô‡∏¢‡∏≥
